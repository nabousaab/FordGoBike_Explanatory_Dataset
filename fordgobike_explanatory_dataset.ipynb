# Cleaning dataset


## Quality issue:
### - We have a lot of empty rows that I do recommend to drop them

```python
bikes_clean = bikes.dropna()
bikes_clean
```

```python
bikes_clean.shape
```

## Tidiness issue
### - We should also drop the columns that have no value to our exploratory and explanatory plotting.

```python
bikes_clean.drop(['start_station_latitude', 'start_station_longitude', 'end_station_latitude', 'end_station_longitude',
                 'start_station_id','end_station_id', 'bike_id' ], axis = 1, inplace = True)
```

```python
bikes_clean.head()
```

## Setting a Categorical status for the start dates as date

```python
from datetime import datetime
bikes_clean['start_time'] = pd.to_datetime(bikes_clean['start_time'])
```

```python
bikes_clean['day'] = bikes_clean['start_time'].apply(lambda x : x.strftime('%A').lower())

days = ['monday', 'tuesday', 'wednesday', 'thursday', 'friday', 'saturday', 'sunday']
dday = pd.api.types.CategoricalDtype(ordered = True, categories = days)

bikes_clean['day'] = bikes_clean['day'].astype(dday);
```

# Univariate plotting
##  Eploratory and Explanatory analysis

```python
plt.figure(figsize = [14, 6])

bins = np.arange(0, bikes_clean['duration_sec'].max()+200, 200)
plt.hist(data = bikes_clean, x = 'duration_sec', bins = bins)
plt.xlim(0, 5000)
plt.title('2019 distribution for Trip Duration (sec)')
plt.xlabel('Duration in sec')
plt.ylabel('Distribution of Frequency');
```

- The distribution is right skewed, showing that most of the trips happen for a period less than half and hor, 
- This suggest to use a log distribution to reconfirm if the hist will change distribution model. 

```python
plt.figure(figsize = [14, 6])
sb.distplot(bikes_clean['duration_sec'], kde = True, hist_kws = {'alpha' : 0.5})
plt.xlim(0, 5000)
plt.title('2019 distribution for Trip Duration (sec)')
plt.xlabel('Duration in sec')
plt.ylabel('Distribution of Frequency');
```

- The distribution plot confirms that the data is skewed toward the half and hour and less, So lets look at the log scale type and reconfirm how the time is been distributed amongst the scale.

```python
np.log10(bikes_clean.describe())
```

```python
plt.figure(figsize = [14, 6])
bins_log = 10 ** np.arange(1.5, 5+0.1, 0.1)
ticks = [100, 300, 1000, 3000, 10000]
labels = ['{}'.format(v) for v in ticks]

plt.hist(data = bikes_clean, x = 'duration_sec', bins = bins_log)
plt.xscale('log')
plt.xticks(ticks, labels)
plt.title('2019 distribution for Trip Duration (sec)')
plt.xlabel('Duration in sec / Limit and Scale changed to log type')
plt.ylabel('Distribution of Frequency');
```

- After Applying the log scale type, the hist shape shows that the data now are almost evenly distributed. And a bit skewed for the less than half hour duration, mainly between 5 and 10 min.
